### Functions to plot difference between two models

```python
def plot_difference(mdiff, title="", annotation=None, ax=None):
    im = ax.imshow(mdiff, cmap='RdBu_r', origin='lower')
    ax.set_title(title)
    fig.colorbar(im, ax=ax)
    return ax
```

### How topics within one model correlate with each other (theory)

In an ideal world, we would like to see different topics decorrelated between themselves. In this case, our matrix would look like this:

```python
num_topics = 10
mdiff = np.ones((num_topics, num_topics))
np.fill_diagonal(mdiff, 0.)

fig, (ax1) = plt.subplots(nrows=1, ncols=1, figsize=(15,7))
plot_difference(mdiff, title="Topic difference (one model) in ideal world", ax=ax1)
plt.tight_layout()
```

- almost red cell - strongly decorrelated topics  
- almost blue cell - strongly correlated topics
